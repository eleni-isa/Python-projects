```python
import pandas as pd
import numpy as np
df = pd.read_csv('dataset_sample.csv') #default: separated with commas
df #data set is printed every time to show the results
```

```python
df.shape #shows the shape of the Data Frame (rows*columns)
```

```python
df.columns
```

```python
df = df.drop_duplicates(keep='first') #drop line 1 'cause was a duplicate of row 0
df
```

```python
#Option 1: fill the NaN values with valuable values
new_df = df.replace({-999: np.nan}) #replace value=-999 in row 8 with NaN
new_df
```

```python
new_df = new_df.fillna({'day':10/10/2018,
                   'city': 'London',
                   'weather': 'rain',
                   'temperature': 15.0,
                   'windspeed':3.0,
                   'feelslike':11.0,
                   'goingout':'yes'}) #fill the NaN cells with specific values
new_df
```

```python
#Option 2: drop the NaN values
new_df2 = df.dropna(how='all')
new_df2
```

```python
new_df2 = new_df2.dropna(axis='columns', thresh=3)
new_df2
```
